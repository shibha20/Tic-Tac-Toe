```python
from IPython.display import clear_output

#print the board
def display_board(board):
    print ('--------------------------------------------------')
    print(str(board[0]) +'  |  '+str(board[1])+ '  |  '+str(board[2]))
    print ('--------------------------------------------------')
    print(str(board[3]) +'  |  '+str(board[4])+ '  |  '+str(board[5]))
    print ('--------------------------------------------------')
    print(str(board[6]) +'  |  '+str(board[7])+ '  |  '+str(board[8]))
    print ('--------------------------------------------------')
    
    
def position_description():
    print ('------')
    print(str(1) +'|'+str(2)+ '|'+str(3))
    print ('------')
    print(str(4) +'|'+str(5)+ '|'+str(6))
    print ('------')
    print(str(7) +'|'+str(8)+ '|'+str(9))
    print ('------')
        


    
```

```python
 def players_choice():
    rightmark = False
    mark = ['']*2
    ##validate whether the input is X or O and is a string
    while rightmark == False:
        ourmark= input("You are Player 1. Please pick a mark 'X' or 'O':  ")
        clear_output()
        
        ##let the first player choose 'X' or 'O'
        if (ourmark.capitalize()  == 'X' and isinstance(ourmark, str) == True):
            
            ##If first player chooses 'X', second player is assigned to 'O'.
            mark[0] = 'X'
            mark[1] = 'O'
            print ("Player 1 chose " + str(mark[0]) +"."+ "Player 2 is assigned to "+ str(mark[1]) + ".")
            print ("")
            rightmark = True 
        elif (ourmark.capitalize()  == 'O' and isinstance(ourmark, str) == True):
            
            ##If first player chooses 'O', second player is assigned to 'X'.
            mark[0] = 'O'
            mark[1] = 'X'
            print ("Player 1 chose " + str(mark[0]) +"."+ "Player 2 is assigned to  "+ str(mark[1]) + ".")
            print ("")
            #set rightmark to True and exit out of while loop once a right mark is inputted
            rightmark = True
    else:
        rightmark= False
    return mark 

```

```python
#let user input where in the board player wants to put X or O
def position(board):
    rightpos = False  
    #ask for input until a right index/integer is picked
    while rightpos == False:
        pos=  int(input("Please pick a number from 1-9: "))
        
        #check whether the inputted position is in range and there is an empty space in the position inputted
        if pos >= 1 and pos <=9 and space_check((pos-1),board)== True:
            pos = pos-1
            rightpos= True
        else: 
            print ("The position is already filled or it is out of range.")
            rightpos = False 
    return pos

#check if where the player wants to put mark is empty
def space_check(pos, board):
    return board[pos] == '             '

#insert the inputted mark to wherever the player wants X or O to be 
def place_marker(board,ourmark,pos):
    space= space_check(pos,board)
    if space== True: 
        board [pos] = str('          ')+str(ourmark.capitalize())

#check if the player won after the last input and return False to indicate game_on be set to be False 
def win_check(board,mark,pos):
    #check first row 
    if (str(board[0]) == str(board [1]) == str(board[2]) == str(board[pos]) or 
    (str(board[3]) == str(board [4]) == str(board[5]) == str(board[pos])) or
    (str(board[6]) == str(board [7]) == str(board[8]) == str(board[pos])) or 
    (str(board[0]) == str(board [3]) == str(board[6]) == str(board[pos])) or  
    (str(board[1]) == str(board [4]) == str(board[7]) == str(board[pos])) or  
    (str(board[2]) == str(board [5]) == str(board[8]) == str(board[pos])) or  
    (str(board[0]) == str(board [4]) == str(board[8]) == str(board[pos])) or 
    str(board[2]) == str(board [4]) == str(board[6]) == str(board[pos])): 
        print ("You have won.")
        return False 
    ##if player has not won yet, check to see if the board is Full and return False to indicate game_on be set to be False
    elif (full_board_check(board) == True):
        print ("The board is full.")
        return False 
    else: 
        return True
         
```
